```python
import pandas as pd
```

```python
df2000 = pd.read_csv('2000_2009 County Population.csv',encoding='latin-1')
df2000.head()
```

```python
df2010 = pd.read_csv('2010-2019 County Population.csv')
df2010['County'] = df2010['County'].str[1:]
df2010.head()
```

```python
df2020 = pd.read_csv('2020-2022 County Population.csv')
df2020['County'] = df2020['County'].str[1:]
df2020.head()
```

```python
df2010_merged = df2010.merge(df2020, how = 'outer', left_on = 'County', right_on = 'County', )
df2010_merged.head()
```

we will inevidebly throw out alaska as it's an outlier from the continental states. A big concern is connecticut, who changed from 'counties' to planning regions in 2020. They don't match up (8 counties vs 9 planning regions). Going to leave out the 2020 data for connecticut. 

```python
df2010_merged[df2010_merged['2010'].isna()]
```

```python
df2010_merged[df2010_merged['2020'].isna()]
```

```python
county_names = pd.DataFrame(df2010_merged['County'].str.split(', ').values.tolist())
df2010_merged = pd.concat([df2010_merged, county_names[[0,1]].rename(columns= {0:'County_name', 1:'State_name'})], axis=1)
```

```python
df2010_merged
```

```python
population_final = df2000.merge(df2010_merged, how = 'outer', left_on= ['CTYNAME','STNAME'], right_on =['County_name', 'State_name'])
```

```python
#note only alaksa and connecticut don't nicely join on. one issue for south dakota, north dakota, louisiana

population_final[population_final['STNAME'].isna()]
```

```python
#removing rows without fips code
population_final.dropna(axis=0, subset=['STATE']).to_csv('County_population_2000_2022.csv')
```

```python
#checking how many rows contain a nan
population_final.shape[0] - population_final.dropna().shape[0]
```

```python
population_final[population_final.isna().any(axis=1)].to_csv('population_NA.csv')
```
