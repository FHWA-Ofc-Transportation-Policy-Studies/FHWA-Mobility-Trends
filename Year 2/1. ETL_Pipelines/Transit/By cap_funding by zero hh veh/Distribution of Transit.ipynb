## Run this notebook second when creating the UPT/transit indicator. It creates the distribution

```python
import pandas as pd
```

```python
#read in ACS rider weights
rider_tms = pd.read_csv('households without vehicle by county.csv').dropna()
rider_tms
```

```python
rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(int)
rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(str)

FIPS_county, FIPS_state = [], []

for i in range(len(rider_tms)):
    fip_code = rider_tms.iloc[i]['FIPS_CODE']
    if len(fip_code) == 5:
        FIPS_county.append(fip_code[2:])
        FIPS_state.append(fip_code[0:2])
    elif len(fip_code) == 4:
        FIPS_county.append(fip_code[-3:])
        FIPS_state.append('0' + fip_code[0])

rider_tms['STATE_CODE'] = FIPS_state
rider_tms['COUNTY_CODE'] = FIPS_county

rider_tms['STATE_CODE'] = rider_tms['STATE_CODE'].astype(int)
rider_tms['COUNTY_CODE'] = rider_tms['COUNTY_CODE'].astype(int)
rider_tms['YEAR'] = rider_tms['YEAR'].astype(str)
rider_tms = rider_tms[['STATE_CODE', 'COUNTY_CODE', 'FIPS_CODE', 'YEAR', 'ZERO_CAR_HOUSEHOLDS']]

rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(float)

rider_tms
```

```python
UPT_count = pd.read_csv('UPT_2002_2023_final_capital funding.csv').iloc[:,1:].drop_duplicates()
UPT_count = UPT_count.drop(columns = [ '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999',
       '2000', '2001'])
unique_fips_uza = UPT_count[['CountyFIPS', 'UZA_CODE']].drop_duplicates()
unique_fips_uza
```

```python
unique_fips_uza[unique_fips_uza['CountyFIPS'] == 6059]
```

```python
UPT_count
```

```python
UPT_count[UPT_count['UZA'] == 2]
```

```python
#join uza code
rider_tms_fips = rider_tms.merge(unique_fips_uza, how = 'left', left_on = 'FIPS_CODE', right_on = 'CountyFIPS')
#sum by UZA
sum_table = rider_tms_fips[['UZA_CODE', 'ZERO_CAR_HOUSEHOLDS', 'YEAR']].dropna().groupby(['UZA_CODE','YEAR']).sum().reset_index()
rider_tms_fips = rider_tms_fips.merge(sum_table.rename(columns = {'ZERO_CAR_HOUSEHOLDS':'SUM_HOUSEHOLDS'}),
                                      how = 'left', left_on = ['UZA_CODE','YEAR'], right_on = ['UZA_CODE','YEAR'])
rider_tms_fips['percent_dist'] = rider_tms_fips['ZERO_CAR_HOUSEHOLDS'] / rider_tms_fips['SUM_HOUSEHOLDS']
rider_tms_fips['percent_dist'] = rider_tms_fips['percent_dist'].fillna(1)
rider_tms_fips.sort_values(by=['STATE_CODE', 'COUNTY_CODE'])
```

```python
rider_tms_fips[rider_tms_fips['FIPS_CODE'] == 6059]
```

```python
UPT_count.columns = ['UZA', '2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022', 'CountyFIPS', 'UZA_CODE']

UPT_count_melt = pd.melt(UPT_count, id_vars = ['UZA','UZA_CODE', 'CountyFIPS'],
                         value_vars = ['2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022']).rename(columns={'variable':'YEAR', 'value':'UPT', 'CountyFIPS':'FIPS_CODE'})
UPT_count_melt = UPT_count_melt[['YEAR', 'UPT', 'UZA_CODE']]
UPT_count_melt
```

```python
UPT_count_melt[UPT_count_melt['UZA_CODE'] == 2]
```

```python
UPT_final = UPT_count_melt.merge(rider_tms_fips, how = 'left', left_on = ['UZA_CODE', 'YEAR'], right_on= ['UZA_CODE', 'YEAR'])
```

```python
UPT_final['distributed_UPT'] = UPT_final['UPT'] * UPT_final['percent_dist']
UPT_final
```

```python
UPT_final[UPT_final['FIPS_CODE'] == 6059]
```

```python
UPT_final = UPT_final.dropna()
```

```python
UPT_final_filtered = UPT_final[UPT_final['YEAR'] != '2002']
UPT_final_filtered = UPT_final_filtered[UPT_final_filtered['YEAR'] != '2003']
UPT_final_filtered = UPT_final_filtered[UPT_final_filtered['YEAR'] != '2004']
```

```python
UPT_final_filtered[UPT_final_filtered['FIPS_CODE'] == 34003]
```

```python
UPT_final_filtered.drop_duplicates()
```

```python
UPT_final_filtered = UPT_final_filtered.groupby(['YEAR', 'STATE_CODE', 'COUNTY_CODE', 'FIPS_CODE']).sum('distributed_UPT').reset_index()
```

```python
UPT_final_filtered[['YEAR', 'STATE_CODE', 'COUNTY_CODE','FIPS_CODE', 'distributed_UPT']].rename(columns = {'distributed_UPT':'Capital Funding'}).drop_duplicates().to_csv('distributed_capital_funding_data by zero_car_households1_9.csv', index = False)
```
