```python
import pandas as pd
```

```python
UPT_df = pd.read_csv('UPT_Cal_Year.csv')
cxw = pd.read_csv('FTA_UZA_County_Xwalk_Ref.csv')
```

```python
UPT_df.head()
```

```python
UPT_df.dtypes
```

```python
#clean and sum across
UPT_df_stg = UPT_df.drop(columns = ['NTD ID','Legacy NTD ID','Agency','UZA Name','Reporter Type','Status', 'TOS', 'Mode'])
UPT_df_stg = UPT_df_stg.groupby(['UZA']).sum()
UPT_df_stg.reset_index(inplace=True)
```

```python
cxw.head()
```

```python
#adjusting cxw data
df1 = cxw[['CountyFIPS','1FTA_UZA_Code']].rename(columns={'1FTA_UZA_Code': 'UZA_CODE'})
df2 = cxw[['CountyFIPS','2FTA_UZA_Code']].rename(columns={'2FTA_UZA_Code': 'UZA_CODE'})
df3 = cxw[['CountyFIPS','3FTA_UZA_Code']].rename(columns={'3FTA_UZA_Code': 'UZA_CODE'})
df4 = cxw[['CountyFIPS','4FTA_UZA_Code']].rename(columns={'4FTA_UZA_Code': 'UZA_CODE'})
FIPS_UZA = pd.concat([df1, df2, df3, df4]).dropna()
FIPS_UZA
```

```python
UPT_df = UPT_df_stg.merge(FIPS_UZA, how='left', left_on = 'UZA', right_on = 'UZA_CODE')
```

```python
UPT_df
```

```python
UPT_df[['CountyFIPS','UZA_CODE']].groupby([ 'CountyFIPS']).count().reset_index().sort_values(by='UZA_CODE', ascending=False)
```

```python
UPT_df.dropna().to_csv('UPT_2002_2023_final.csv')
```
