## Run this notebook second when creating the UPT/transit indicator. It creates the distribution

```python
import pandas as pd
```

```python
#read in ACS rider weights
rider_tms = pd.read_csv('TMS_count_of_Riders.csv').dropna()
rider_tms
```

```python
rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(int)
rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(str)

FIPS_county, FIPS_state = [], []

for i in range(len(rider_tms)):
    fip_code = rider_tms.iloc[i]['FIPS_CODE']
    if len(fip_code) == 5:
        FIPS_county.append(fip_code[2:])
        FIPS_state.append(fip_code[0:2])
    elif len(fip_code) == 4:
        FIPS_county.append(fip_code[-3:])
        FIPS_state.append('0' + fip_code[0])

rider_tms['STATE_CODE'] = FIPS_state
rider_tms['COUNTY_CODE'] = FIPS_county

rider_tms['STATE_CODE'] = rider_tms['STATE_CODE'].astype(int)
rider_tms['COUNTY_CODE'] = rider_tms['COUNTY_CODE'].astype(int)
rider_tms['YEAR'] = rider_tms['YEAR'].astype(str)
rider_tms = rider_tms[['STATE_CODE', 'COUNTY_CODE', 'FIPS_CODE', 'YEAR', 'TMS_COUNT']]

rider_tms['FIPS_CODE'] = rider_tms['FIPS_CODE'].astype(float)

rider_tms
```

```python
UPT_count = pd.read_csv('UPT_2002_2023_final.csv').iloc[:,1:]
unique_fips_uza = UPT_count[['CountyFIPS', 'UZA_CODE']].drop_duplicates()
unique_fips_uza
```

```python
UPT_count
```

```python
#join uza code
rider_tms_fips = rider_tms.merge(unique_fips_uza, how = 'left', left_on = 'FIPS_CODE', right_on = 'CountyFIPS')
sum_table = rider_tms_fips[['UZA_CODE', 'TMS_COUNT', 'YEAR']].dropna().groupby(['UZA_CODE','YEAR']).sum().reset_index()
rider_tms_fips = rider_tms_fips.merge(sum_table.rename(columns = {'TMS_COUNT':'SUM_TMS'}),
                                      how = 'left', left_on = ['UZA_CODE','YEAR'], right_on = ['UZA_CODE','YEAR'])
rider_tms_fips['percent_dist'] = rider_tms_fips['TMS_COUNT'] / rider_tms_fips['SUM_TMS']
rider_tms_fips['percent_dist'] = rider_tms_fips['percent_dist'].fillna(1)
rider_tms_fips.sort_values(by=['STATE_CODE', 'COUNTY_CODE'])
```

```python
rider_tms_fips.dtypes
```

```python
UPT_count.columns = ['UZA', '2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022', '2023', 'CountyFIPS', 'UZA_CODE']

UPT_count_melt = pd.melt(UPT_count, id_vars = ['UZA','UZA_CODE', 'CountyFIPS'],
                         value_vars = ['2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022', '2023']).rename(columns={'variable':'YEAR', 'value':'UPT', 'CountyFIPS':'FIPS_CODE'})
UPT_count_melt = UPT_count_melt[['YEAR', 'UPT', 'FIPS_CODE']]
UPT_count_melt
```

```python
UPT_count_melt.dtypes
```

```python
UPT_final = UPT_count_melt.merge(rider_tms_fips, how = 'left', left_on = ['FIPS_CODE', 'YEAR'], right_on= ['FIPS_CODE', 'YEAR'])
UPT_final
```

```python
UPT_final['distributed_UPT'] = UPT_final['UPT'] * UPT_final['percent_dist']
UPT_final
```

```python
x1 = UPT_final[UPT_final['FIPS_CODE'] == 6083]
x1[x1['YEAR'] == '2012']
```

```python
UPT_final_filtered = UPT_final[UPT_final['YEAR'] != '2002']
UPT_final_filtered = UPT_final_filtered[UPT_final_filtered['YEAR'] != '2003']
UPT_final_filtered = UPT_final_filtered[UPT_final_filtered['YEAR'] != '2004']
```

```python
UPT_final_filtered = UPT_final_filtered[['YEAR', 'STATE_CODE', 'COUNTY_CODE','distributed_UPT']]
```

```python
UPT_final_filtered = UPT_final_filtered.groupby(['YEAR', 'STATE_CODE', 'COUNTY_CODE']).sum('distributed_UPT').reset_index()
```
