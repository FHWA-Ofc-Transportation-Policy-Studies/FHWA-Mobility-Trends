```python
def update_lnmile_data(df,year):
    df['YEAR'] = year
    
    df1 = df[['State_Cd', 'COUNTY_CD', 'lnmiles', 'YEAR']].rename(
        columns={'State_Cd': 'STATE', 'COUNTY_CD': 'COUNTY', 'lnmiles':'LNMILES'})
    df2 = df[['State_Cd', 'COUNTY_CD', 'loclnmiles', 'YEAR']].rename(
        columns={'State_Cd': 'STATE', 'COUNTY_CD': 'COUNTY', 'loclnmiles':'LNMILES'})
    df3 = df[['State_Cd', 'COUNTY_CD', 'total lnmiles', 'YEAR']].rename(
        columns={'State_Cd': 'STATE', 'COUNTY_CD': 'COUNTY', 'total lnmiles':'LNMILES'})
    
    df1['Data_cat'] = 1
    df2['Data_cat'] = 2
    df3['Data_cat'] = 3

    return pd.concat([df1, df2, df3], ignore_index = True)
    
```

```python
for i in range(2010,2020):
    df_stg = pd.read_excel(xls, str(i))
    concat_df = update_lnmile_data(df_stg, i)
    df = pd.concat([df, concat_df], ignore_index = True, axis=0)
```

```python
def check_state_total_for_year(df, state_fips, final_df, year, state):
    #cut down main df to year and state. Get sum of year/state miles
    df = df[df['YEAR'] == year]
    df = df[df['STATE'] == state]
    grouped_df = df[['LNMILES', 'Data_cat']].groupby(by=['Data_cat']).sum().reset_index()
    
    #get the correct milage
    state_miles = state_fips[state_fips['STATE_CODE'] == state].iloc[0]['MILES']
    
    #index the best data cat
    ln_miles_list = list(grouped_df['LNMILES'])
    index = min(range(len(ln_miles_list)), key=lambda i: abs(ln_miles_list[i]-state_miles))
    final_df_stg = df[df['Data_cat'] == grouped_df['Data_cat'].iloc[index]]
    final_df = pd.concat([final_df,final_df_stg], ignore_index = True)
    
    return final_df
```

```python
final_df = pd.DataFrame()
failure_list = []
state_code_list = list(state_fips['STATE_CODE'])

for i in range(2000, 2021):
    print(i)
    state_file = pd.ExcelFile('state files/'+ str(i) + '.xls')
    state_df= pd.read_excel(state_file, 'A', skiprows=13)
    state_fips = state_df.iloc[:, [0, -1]]
    state_fips.columns = ['STATE_NAME', 'MILES']
    
    state_fips_codes = pd.read_csv('state files/state_fips_codes.csv')
    state_fips_codes = state_fips_codes[['st']]
    state_fips = pd.concat([state_fips, state_fips_codes], axis = 1)
    state_fips = state_fips[['STATE_NAME', 'st', 'MILES']].rename(columns = {'st': 'STATE_CODE'})
    state_fips['STATE_CODE'] = state_fips['STATE_CODE'].astype(float)
    
    for j in tqdm(state_code_list):
        try:
            final_df = check_state_total_for_year(df, state_fips, final_df, i, j)
        except:
            failure_list.append([i, j])



    
```
