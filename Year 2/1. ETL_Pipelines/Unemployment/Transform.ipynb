## Builds the Unemployment variable, which is simply a concatenation of each unemployment file.

Source: https://www.bls.gov/lau/tables.htm
        
State data is an aggregation of the county level data

**Unemployment by county**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('laucnty00.xlsx')
df.head()
```

```python
df1 = df.iloc[5:]
df1.columns = ['Code', 'State_Code', 'County_Code', 'County_name', 'Year', 
               'None', 'Labor_Force' , 'Employed', 'Unemployed', 'Unemployment_Rate']

for i in range(1, 22):
    if len(str(i)) == 1:
        year_code = '0' + str(i)
    else:
        year_code = str(i)
        
    stg_df = pd.read_excel('laucnty' + year_code + '.xlsx')
    stg_df = stg_df.iloc[5:]
    stg_df.columns = df1.columns
    df1 = pd.concat([df1, stg_df], ignore_index = True)

        
```

```python
df1.to_csv('Final_Unemployment_County.csv')
```
