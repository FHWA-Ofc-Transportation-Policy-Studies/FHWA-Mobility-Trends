```python
import pandas as pd
import numpy as np
```

```python
#Pulling in all of the 2016-2020 data, then moving down to just Interstate and other principal arterial

df = pd.read_csv('Concat_data_2016_2020.csv')
df = df.apply(pd.to_numeric, errors = 'coerce').fillna(0)
df
```

```python
df['VMT_Total_daily'] = df['RINT'] + df['ROPA'] + df['UINT'] + df['UOPA']
vmt1 = df[['ST', 'CO','YEAR', 'VMT_Total_daily']]
#account for leap year in multiplication of daily values
vmt1.loc[vmt1['YEAR'].isin([2016,2020]), 'VMT_Total'] = vmt1['VMT_Total_daily'] * 366
vmt1.loc[vmt1['YEAR'].isin([2017,2018, 2019]), 'VMT_Total'] = vmt1['VMT_Total_daily'] * 365
vmt1
```

```python
#check DC
#Update DC
df_dc = vmt1[vmt1['ST'] == 11]
df_dc_grouped = df_dc.drop(columns=['VMT_Total_daily']).groupby(['ST', 'YEAR']).sum(['VMT_Total']).reset_index()
df_dc_grouped = df_dc_grouped[['ST', 'CO','YEAR','VMT_Total', ]]
df_dc_grouped['CO'] = 1
df_dc_grouped

vmt1 = vmt1.drop(columns= ['VMT_Total_daily'])
vmt1 = vmt1[vmt1['ST'] != 11]
vmt1 = pd.concat((vmt1, df_dc_grouped),)
vmt1
```

```python
#remove puerto rice and move to csv
vmt1 = vmt1[vmt1['ST'] != 72]
vmt1[['ST', 'CO', 'YEAR','VMT_Total']].to_csv('County_VMT_2016_2020_INT_and_PrinAterial.csv' , index = False)
```
