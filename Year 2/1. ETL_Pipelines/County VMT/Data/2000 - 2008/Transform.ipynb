```python
import pandas as pd
import numpy as np
```

This code file takes the data to only Rural Interstate, Urban Interstate, Rural Other Principal Artieral, and Urban Other Principal Arteral

```python
df = pd.read_csv('Concat_2000_2008_vmt.csv')
df = df[df['F_SYSTEM'] != 12]
df_grouped = df[['YEAR', 'STATE_CODE', 'COUNTY_CODE', 'VMT_TOTAL']].groupby(['YEAR', 'STATE_CODE', 'COUNTY_CODE']).sum('VMT_TOTAL').reset_index()
df_grouped
```

```python
#update dc
df_dc = df_grouped[df_grouped['STATE_CODE'] == 11]
df_dc_grouped = df_dc.groupby(['STATE_CODE', 'YEAR']).sum(['VMT_TOTAL']).reset_index()
df_dc_grouped = df_dc_grouped[['YEAR','STATE_CODE', 'COUNTY_CODE','VMT_TOTAL' ]]
df_dc_grouped['COUNTY_CODE'] = 1
df_dc_grouped

df_grouped = df_grouped[df_grouped['STATE_CODE'] != 11]
df_grouped = pd.concat((df_grouped, df_dc_grouped),)
df_grouped
```

```python
df_grouped.to_csv('COUNTY_VMT_2000_2008.csv', index = False)
```
