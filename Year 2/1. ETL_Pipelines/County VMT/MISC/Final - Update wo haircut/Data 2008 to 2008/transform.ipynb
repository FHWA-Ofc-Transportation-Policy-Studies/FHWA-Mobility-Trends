```python
import pandas as pd
```

```python
df = pd.read_csv('Concat_2000_2008_vmt.csv')
df = df[df['STATE_CODE'] != 72]
df
```

```python
df_grouped = df.drop(columns = ['F_SYSTEM', 'AVMT (1,000)']).groupby(['YEAR','STATE_CODE', 'COUNTY_CODE']).sum('VMT_TOTAL').reset_index()
df_grouped
```

```python
df_annual = df_grouped[['YEAR','VMT_TOTAL']].groupby('YEAR').sum('VMT_TOTAL')
df_annual
```

```python
df_grouped.to_csv('Transformed_2000_2008.csv', index= False)
```
