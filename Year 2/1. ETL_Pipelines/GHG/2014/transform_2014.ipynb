```python
import pandas as pd
```

```python
df = pd.read_csv('onroad_123.csv')
df['pollutant_desc'].unique()
```

```python
def sum_pollutants(df_title):
    df123 = pd.read_csv(df_title)
    df123 = df123[['state_and_county_fips_code','pollutant_desc', 'total_emissions']]
    df123 = df123[df123['pollutant_desc'].isin(['Methane','Carbon Dioxide'])]
    df123_summed = df123.groupby(['state_and_county_fips_code']).sum('total_emissions')
    df123_summed = df123_summed.reset_index()
    
    return df123_summed

df_title_list = ['onroad_123.csv','onroad_4.csv','onroad_5.csv','onroad_67.csv', 'onroad_8910.csv']


final_df = None
for i in df_title_list:
    df_summed = sum_pollutants(i)
    if final_df is None:
        final_df = df_summed
    else:
        final_df = pd.concat([final_df, df_summed], axis = 0)
```

```python
final_df['YEAR'] = 2014
final_df.head()
```

```python
len(final_df)
```

```python
final_df.groupby('state_and_county_fips_code').count().sort_values(by='total_emissions', ascending = False)
```

```python
final_df.to_csv('COUNTY_GHG_2014.csv', index = False)
```
