## This script concatenates the GHG numbers used with the NEI Onroad estimates.

```python
import pandas as pd
```

```python
df = pd.read_csv('2008/COUNTY_GHG_2008.csv')
df
```

```python
df_title_list = ['2008/COUNTY_GHG_2008.csv', '2011/COUNTY_GHG_2011.csv', '2014/COUNTY_GHG_2014.csv',
                 '2017/COUNTY_GHG_2017.csv', '2020/COUNTY_GHG_2020.csv']

final_df = None
for i in df_title_list:
    df = pd.read_csv(i)
    df.columns = ['FIPS_CODE', 'TOTAL_EMISSIONS', 'YEAR']
    final_df = pd.concat([final_df, df])
```

```python
final_df
```

```python
final_df.to_csv('GHG_COUNTY_ALL.csv', index = False)
```
