```python
import pandas as pd
import numpy as np
```

VMT imputation

```python
df = pd.read_csv('VMT_MODELING_DATASET7_25.csv')
df_09_state = pd.read_csv('STATE_VMT_2009.csv')
df_10_state = pd.read_csv('STATE_VMT_2010.csv')
df_fips_state = pd.read_csv('us-state-ansi-fips.csv')
```

```python
df_2008 = df[df['YEAR'] == 2008][['STATE_CODE', 'COUNTY_CODE', 'YEAR', 'VMT']]

df_2008_state = df_2008[['STATE_CODE', 'VMT']].groupby('STATE_CODE').sum('VMT').reset_index().rename(columns = {'VMT':'STATE_VMT'})
df_2008 = df_2008.merge(df_2008_state, left_on = 'STATE_CODE', right_on = 'STATE_CODE')
df_2008['Percent_Ratio'] = df_2008['VMT'] / df_2008['STATE_VMT'] 

df_2008
```

```python
df_09_state['RURAL INTERSTATE'] = pd.to_numeric(df_09_state['RURAL INTERSTATE'].replace(',','', regex=True), errors='coerce')
df_09_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'] = pd.to_numeric(df_09_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'].replace(',','', regex=True), errors='coerce')
df_09_state[ 'URBAN INTERSTATE'] = pd.to_numeric(df_09_state['URBAN INTERSTATE'].replace(',','', regex=True), errors='coerce')
df_09_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'] = pd.to_numeric(df_09_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'].replace(',','', regex=True), errors='coerce')
df_09_state['URBAN OTHER PRINCIPAL ARTERIAL'] = pd.to_numeric(df_09_state['URBAN OTHER PRINCIPAL ARTERIAL'].replace(',','', regex=True), errors='coerce')

df_09_state = df_09_state.fillna(0)

df_10_state['RURAL INTERSTATE'] = pd.to_numeric(df_10_state['RURAL INTERSTATE'].replace(',','', regex=True), errors='coerce')
df_10_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'] = pd.to_numeric(df_10_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'].replace(',','', regex=True), errors='coerce')
df_10_state[ 'URBAN INTERSTATE'] = pd.to_numeric(df_10_state['URBAN INTERSTATE'].replace(',','', regex=True), errors='coerce')
df_10_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'] = pd.to_numeric(df_10_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'].replace(',','', regex=True), errors='coerce')
df_10_state['URBAN OTHER PRINCIPAL ARTERIAL'] = pd.to_numeric(df_10_state['URBAN OTHER PRINCIPAL ARTERIAL'].replace(',','', regex=True), errors='coerce')

df_10_state = df_10_state.fillna(0)
```

```python
df_09_state = df_09_state.rename(columns = {'Unnamed: 0':'State'})
df_10_state = df_10_state.rename(columns = {'Unnamed: 0':'State'})

df_09_state = df_09_state[['State', 'RURAL INTERSTATE',
       'RURAL OTHER FREEWAYS AND EXPRESSWAYS','URBAN INTERSTATE',
       'URBAN OTHER FREEWAYS AND EXPRESSWAYS',
       'URBAN OTHER PRINCIPAL ARTERIAL',]]

df_09_state['VMT_SUM'] = df_09_state['RURAL INTERSTATE']+ \
                            df_09_state['URBAN INTERSTATE'] + \
                            df_09_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'] + \
                            df_09_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'] + \
                            df_09_state['URBAN OTHER PRINCIPAL ARTERIAL']


df_10_state = df_10_state[['State', 'RURAL INTERSTATE',
       'RURAL OTHER FREEWAYS AND EXPRESSWAYS','URBAN INTERSTATE',
       'URBAN OTHER FREEWAYS AND EXPRESSWAYS',
       'URBAN OTHER PRINCIPAL ARTERIAL',]]

df_10_state['VMT_SUM'] = df_10_state['RURAL INTERSTATE'] + df_10_state['URBAN INTERSTATE'] + df_10_state['RURAL OTHER FREEWAYS AND EXPRESSWAYS'] + \
                        df_10_state['URBAN OTHER FREEWAYS AND EXPRESSWAYS'] + df_10_state['URBAN OTHER PRINCIPAL ARTERIAL']
                    
```

```python
df_state_totals = df_09_state[['State', 'VMT_SUM']].merge(df_10_state[['State', 'VMT_SUM']], left_on = 'State', right_on = 'State')
df_state_totals = df_state_totals.rename(columns = {'VMT_SUM_x': 'VMT_2009', 'VMT_SUM_y':'VMT_2010'})
df_state_totals = df_state_totals.replace('Dist. of Columbia', 'District of Columbia')
df_state_totals = df_state_totals.merge(df_fips_state, left_on = 'State', right_on = 'stname')
df_2008 = df_2008.merge(df_state_totals, left_on = 'STATE_CODE', right_on = ' st')
df_2008['2009'] = df_2008['VMT_2009'] * df_2008['Percent_Ratio'] * 1000000 #state numbers are expressed in millions
df_2008['2010'] = df_2008['VMT_2010'] * df_2008['Percent_Ratio'] * 1000000
```

```python
df_county_data = df_2008[['STATE_CODE', 'COUNTY_CODE', '2009', '2010']]
df_county_data = pd.melt(df_county_data, id_vars = ['STATE_CODE', 'COUNTY_CODE'], value_vars = ['2009', '2010']).rename(columns = {'variable':'imputed_year', 'value': 'imputed_VMT'})
df_county_data['imputed_year'] = df_county_data['imputed_year'].astype(float)
df = df.merge(df_county_data, how = 'left', left_on = ['STATE_CODE', 'COUNTY_CODE', 'YEAR'], right_on = ['STATE_CODE', 'COUNTY_CODE', 'imputed_year'])
df['VMT'] = df['VMT'].fillna(df['imputed_VMT'])
```

Lane miles

```python
df_lnmiles = df[df['YEAR'].isin([2008,2011])][['STATE_CODE', 'COUNTY_CODE', 'YEAR', 'LNMILES']]
df_lnmiles_2008 = df_lnmiles[df_lnmiles['YEAR'] == 2008].rename(columns = {'LNMILES': 'LNMILES_2008'})
df_lnmiles_2010 = df_lnmiles[df_lnmiles['YEAR'] == 2010].rename(columns = {'LNMILES': 'LNMILES_2010'})

df_lnmiles = df_lnmiles_2008.merge(df_lnmiles_2010, how = 'left', left_on = ['STATE_CODE', 'COUNTY_CODE'], right_on = ['STATE_CODE', 'COUNTY_CODE'])
df_lnmiles['2009'] = (df_lnmiles['LNMILES_2010'] - df_lnmiles['LNMILES_2008']) * (1/2) +  df_lnmiles['LNMILES_2008']

df_lnmiles = df_lnmiles[['STATE_CODE', 'COUNTY_CODE', '2009']]
df_lnmiles = df_lnmiles.melt(id_vars = ['STATE_CODE', 'COUNTY_CODE'], value_vars = ['2009']).rename(columns = {'variable':'imputed_YEAR', 'value':'imputed_LNMILES'})
df_lnmiles
```

```python
df.dtypes, df_lnmiles.dtypes
```

```python
df_lnmiles['imputed_YEAR'] = df_lnmiles['imputed_YEAR'].astype(float)
df = df.merge(df_lnmiles, how = 'left', left_on = ['STATE_CODE', 'COUNTY_CODE', 'YEAR'], right_on = ['STATE_CODE', 'COUNTY_CODE', 'imputed_YEAR'])
df['LNMILES'] = df['LNMILES'].fillna(df['imputed_LNMILES'])
```

```python
df['POPULATION'] = pd.to_numeric(df['POPULATION'].replace(',','', regex=True), errors='coerce')
df['Unemployment_Rate'] = pd.to_numeric(df['Unemployment_Rate'].replace(',','', regex=True), errors='coerce')
```
