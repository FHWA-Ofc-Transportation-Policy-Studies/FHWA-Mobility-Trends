```python
#write function that plots any two columns and colors the points by geo type 

def scatter(data, v1, v2, v3, chart_title, df2):
    #assign the x and y axis limits to be the max of the indicator across all geotypes 
    xmax = max(df2[str(v1)])
    ymax = max(df2[str(v2)])
    
    fg = sns.FacetGrid(data=data, hue=v3, aspect=1.61, palette="muted")
    fg = fg.map(plt.scatter, v1, v2, alpha = 0.5).add_legend()
    fg.fig.suptitle(chart_title)
    
    #set x and y axis limits 
    plt.xlim(0, xmax) #set to be the max of the series 
    plt.ylim(0, ymax)
    
plt.show(scatter(df_indicators, v1 = "TELEWORK", v2 = "VMT", v3 = "GEMS_Typo", chart_title = "Geotype = All" , df2 = df_indicators))
plt.show(scatter(df_c, v1 = "TELEWORK", v2 = "VMT", v3 = "GEMS_Typo", chart_title = "Geotype = C \n(Highly Dispersed Monocentric Rural)" , df2 = df_indicators))
```

### Correlation

```python
def corr_matrix(data, chart_title):
    plt.figure(figsize=(15,10))
    sns.heatmap(data.corr().round(2),cmap='Blues',annot=True) 
    plt.title(chart_title, y=1.02)

plt.show(corr_matrix(df_indicators, chart_title = "Correlation Matrix: Geotype All" ))
#plt.show(corr_matrix(df_a, chart_title = "Correlation Matrix: Geotype A (High Polycentric Urban)" ))
```

Here we are interested in which indicators are most and least correlated with VMT

```python
#returns df with the 3 most correlated pairs (largest absolute value correlation not including correlation = 1)
def corr_highest(data):
    corr_df = pd.DataFrame(data.corr().abs().unstack().sort_values().drop_duplicates().reset_index())
    corr_df = corr_df.rename(columns = {'level_0':'Var_1','level_1':'Var_2', 0:'Correlation_Abs_Val'})
    corr_df = corr_df.loc[(corr_df.Correlation_Abs_Val != 1)]
    corr_df = corr_df.loc[(corr_df.Var_1 == 'VMT') |(corr_df.Var_2 == 'VMT') ]
    corr_df_high = corr_df.iloc[-3:].round(4)
    corr_df_high = corr_df_high.sort_values(by = 'Correlation_Abs_Val', ascending = False)
    return corr_df_high

#returns df with the 3 least correlated pairs (lowest absolute value correlation aka closest to zero)
def corr_lowest(data):
    corr_df = pd.DataFrame(data.corr().abs().unstack().sort_values().drop_duplicates().reset_index())
    corr_df = corr_df.rename(columns = {'level_0':'Var_1','level_1':'Var_2', 0:'Correlation_Abs_Val'})
    corr_df = corr_df.loc[(corr_df.Correlation_Abs_Val != 1)]
    corr_df = corr_df.loc[(corr_df.Var_1 == 'VMT') |(corr_df.Var_2 == 'VMT') ]
    corr_df_low = corr_df.iloc[0:3].round(4)
    return corr_df_low
```

### Identifying Extreme Values 

Defining these points as outliers assumes a normal distribution, which is not realistic looking at the distribution charts

```python
#function to find outliers using interquartile ranges (IQR)
#also shows percent of data points that are outliers for that specific indicator & geotype

def find_outliers(df, col):
    geos_list = str(df["GEMS_Typo"].unique())
    df = df[str(col)]
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    print("geotypes included: " + geos_list)
    print("total number of data points: " + str(len(df)))
    print("number of outliers: " + str(len(outliers)))
    outlier_percent = round(len(outliers) / len(df), 2)
    print("percent of data that are outliers: " + str(outlier_percent) )
    print("max outlier value: " + str(outliers.max()))
    print("min outlier value: " + str(outliers.min()))
    print(outliers)
    print("\n")
```

```python
#See how many components to create
for i in range(1,7):
    pca = decomposition.PCA(n_components = i)
    pca.fit(df_scaled)
    print( i, "components explain", round(sum(pca.explained_variance_ratio_), 4), "percent of the variance")
```

Since 3 components explain most (> 90%) of the variance in the data set, we are interested in the highest contributing variables in each of those components

```python
importance_frame = pd.DataFrame(pca.components_,columns=df_scaled.columns,index = ['PC-1','PC-2', 'PC-3'])
importance_frame
```

```python
n = 7
absolute_valued_frame = abs(importance_frame).T
print(absolute_valued_frame['PC-1'].nlargest(n))
print(absolute_valued_frame['PC-2'].nlargest(n))
print(absolute_valued_frame['PC-3'].nlargest(n))

important_variables = set(list(absolute_valued_frame['PC-1'].nlargest(n).index) + \
                      list(absolute_valued_frame['PC-2'].nlargest(n).index) + \
                      list(absolute_valued_frame['PC-3'].nlargest(n).index))
```

```python
from sklearn.cluster import KMeans, DBSCAN, SpectralClustering
from sklearn import metrics
```

```python
df_cluster = df_indicators[['POPULATION', 'GDP', 'Unemployment_Rate',
                            'Charging_Stations','LNMILES', 'TELEWORK', 'distributed_UPT']].dropna()

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_cluster)
df_scaled_cluster = pd.DataFrame(scaled, columns=df_cluster.columns)
```

Starting off with kmeans analysis: I'm really interested in if 6 clusters, which is equivelant to the number of GEMS geocodes, will fit well to the data. Otherwise, I'd be interesetd in clustering can capture the urban rural split, and if so, it is good enough to possibily split the data into chunks based on this analysis. 

First, lets check to see the distances from centroids to determine number of clusters

```python
#setting seed?

inertia_values = []
for i in range(2,11):
    kmeans = KMeans(n_clusters = i, random_state = 0).fit(df_scaled_cluster)
    inertia_values.append(kmeans.inertia_)
```
