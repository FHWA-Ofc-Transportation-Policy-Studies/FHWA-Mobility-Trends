This script starts with the final year 2 task 4 dataset and joins any additional indicators used in task 6 year 2

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datapath = 'C:/Users/hrowe/Documents/FHWA mobility trend report/T4 - Forecasting/Year 2/FHWA_Mobility_Trends/Year 2 County Level/Modeling_Year1_County_level/Data/'
df_master = pd.read_csv(datapath + 'County_Year2_1_10_24.csv', index_col=0) #final year 2 task 4 modeling dataset as of 1/12/24
df_master
```

##### Joining indicator: Number of Households with Zero Vehicles

```python
hh_zero_veh = pd.read_csv('C:/Users/hrowe/Documents/FHWA mobility trend report/T4 - Forecasting/Year 2/FHWA_Mobility_Trends/Year 2 County Level/Data_engineering_Year1_County_level/Transit/By household wo vehicle - using capital funding/households without vehicle by county.csv')
hh_zero_veh = hh_zero_veh[hh_zero_veh['FIPS_CODE'].notna()] #remove cases where FIPS_CODE is NA
hh_zero_veh['FIPS_CODE'] = hh_zero_veh['FIPS_CODE'].astype(int)#.astype(str).str.zfill(5)
hh_zero_veh = hh_zero_veh[['YEAR', 'FIPS_CODE', 'ZERO_CAR_HOUSEHOLDS']]
hh_zero_veh = hh_zero_veh.rename(columns={'ZERO_CAR_HOUSEHOLDS': 'HH_ZERO_VEH','FIPS_CODE': 'Full_FIPS_Code' })

hh_zero_veh
```

```python
df = pd.merge(df_master, hh_zero_veh, on = ['Full_FIPS_Code', 'YEAR'] , how='left')
df
```
