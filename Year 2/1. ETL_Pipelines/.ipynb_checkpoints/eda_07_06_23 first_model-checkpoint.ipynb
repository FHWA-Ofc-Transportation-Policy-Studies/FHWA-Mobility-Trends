### Correlation

```python
def corr_matrix(data, chart_title):
    plt.figure(figsize=(15,10))
    sns.heatmap(data.corr().round(2),cmap='Blues',annot=True) 
    plt.title(chart_title, y=1.02)

plt.show(corr_matrix(df_indicators, chart_title = "Correlation Matrix: Geotype All" ))
plt.show(corr_matrix(df_a, chart_title = "Correlation Matrix: Geotype A" ))
```

### Distribution Plots 

```python
import warnings

def dist_plots(data, col_num, chart_title):
    #remove geotype column (because it in an object, not numeric)
    data = data.drop('GEMS_Typo', axis = 1)
    warnings.filterwarnings("ignore")
    l = data.columns.values
    number_of_columns = col_num
    number_of_rows = int(len(l)-1/number_of_columns)
    plt.figure(figsize=(5*number_of_columns,5*number_of_rows))
    for i in range(0,len(l)):
        plt.subplot(number_of_rows + 1,number_of_columns,i+1)
        sns.distplot(data[l[i]],kde=True) 
    plt.title(chart_title, x = -1.5, y = 2.3)
    
        
plt.show(dist_plots(df_indicators, 4, "Distribution Plots: Geotype All"))
plt.show(dist_plots(df_a, 4, "Distribution Plots: Geotype A"))
```

### Boxplots

```python
def box_plots(data, num_col, chart_title):
    #remove geotype column (because it in an object, not numeric)
    data = data.drop('GEMS_Typo', axis = 1)
    l = data.columns.values
    number_of_columns = num_col 
    number_of_rows = int(len(l)-1/number_of_columns)
    plt.figure(figsize=(4*number_of_columns,5*number_of_rows))
    for i in range(0,len(l)):
        plt.subplot(number_of_rows + 1,number_of_columns,i+1)
        sns.boxplot(data[l[i]]) 
    plt.title(chart_title, x = -1.5, y = 2.3)

plt.show(box_plots(df_indicators, 4, "Box  Plots: Geotype All"))
```

### Principal Component Analysis

```python
#scale the data using min-max method
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_indicators)
df_scaled = pd.DataFrame(scaled, index=df_indicators.index, columns=df_indicators.columns)
df_scaled.head()
```

```python
df_indicators
```

```python
#drop VMT from scaled data frame for an unsupervised analysis 
df_scaled = df_scaled.drop("VMT", axis = 1)
df_scaled.head()
```

```python
#See how many components to create
for i in range(1,7):
    pca = decomposition.PCA(n_components = i)
    pca.fit(df_scaled)
    print( i, "components explain", round(sum(pca.explained_variance_ratio_), 4), "percent of the variance")
```

Since 3 components explain most (> 90%) of the variance in the data set, we are interested in the highest contributing variables in each of those components

```python
pca = decomposition.PCA(n_components = 3)
df_transformed = pca.fit_transform(df_scaled)
```

```python
importance_frame = pd.DataFrame(pca.components_,columns=df_scaled.columns,index = ['PC-1','PC-2', 'PC-3'])
importance_frame
```

```python
n = 7
absolute_valued_frame = abs(importance_frame).T
print(absolute_valued_frame['PC-1'].nlargest(n))
print(absolute_valued_frame['PC-2'].nlargest(n))
print(absolute_valued_frame['PC-3'].nlargest(n))

important_variables = set(list(absolute_valued_frame['PC-1'].nlargest(n).index) + \
                      list(absolute_valued_frame['PC-2'].nlargest(n).index) + \
                      list(absolute_valued_frame['PC-3'].nlargest(n).index))
```

Most variance seems to lie in unemployment rate & telework (which are both percents rather than absolute numbers)
