Using this script to check differences between the modeling and forecast data between 2005-2020. Hoping this will help understand the discrepancy between forecasts

```python
import pandas as pd
pd.set_option('display.max_columns', 500)

import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#modeling data
m_df = pd.read_csv('C:/Users/zapate/Desktop/FHWA_Mobility_Trends/Year 2 County Level/Modeling_Year1_County_level/Data/County_Year2_11_13_23.csv')
#forecasting_data 
f_df = pd.read_csv('C:/Users/zapate/Desktop/FHWA_Mobility_Trends/Year 2 County Level/forecasting/forecasting_data.csv')
f_df = f_df[f_df['YEAR'] < 2021]
f_df
```

```python
#reduce modeling dataset down to just what we model with
m_df = m_df[['Full_FIPS_Code','YEAR','VMT', 'LNMILES', 'POPULATION',  'TELEWORK', 'Charging_Stations', 'True_GDP', 'Unemployment_Rate']].dropna() #purposely excluding upt since not in model
m_sd = m_df[m_df['VMT'] > 0]
m_df 
```

```python
#first lets figure out how many more entries per year forecasting dataset has over modeling
forecast_aggregate = f_df[['YEAR', 'FIPS']].groupby('YEAR').count()
forecast_aggregate
```

```python
modeling_aggregate = m_df[['YEAR', 'Full_FIPS_Code']].groupby('YEAR').count()
modeling_aggregate
```

```python
#take a look at what fips codes don't get modeled
col_names = []
for i in m_df.columns:
    col_names.append(str(i) + '_modeling')
m_df.columns = col_names
m_df['in_model'] = True

m_df.columns
```

```python
m_df = m_df.rename(columns = {'Full_FIPS_Code_modeling': 'FIPS', 'YEAR_modeling':'YEAR'})

df = f_df.merge(m_df, how = 'left', left_on = ['YEAR', 'FIPS'], right_on = ['YEAR', 'FIPS'])
```

```python
df['in_model'] = df['in_model'].fillna(False)
VMT_agg = df[['in_model', 'YEAR',  'VMT_modeling']].groupby(['YEAR', 'in_model']).sum('VMT')
VMT_agg
```

```python
df_2010 = df[df['YEAR'] == 2010]

df_2010_t = df_2010[df_2010['in_model'] == True]
df_2010_t.describe()            
```

```python
df_2010 = df[df['YEAR'] == 2010]

df_2010_f = df_2010[df_2010['in_model'] == False]
df_2010_f.describe()            
```

```python
#now, we can work on spot checking if all of Matt's forecast data is exactly like the Modeling data
df_2010_t
```

```python
print(df_2010_t['EV Charging Scenario 1'].equals(df_2010_t['Charging_Stations_modeling']))
print(df_2010_t['Lane Miles Baseline'].equals(df_2010_t['LNMILES_modeling']))
print(df_2010_t['Pop'].equals(df_2010_t['POPULATION_modeling']))
print(df_2010_t['Telework Baseline'].equals(df_2010_t['TELEWORK_modeling']))
print(df_2010_t['EIA REAL'].equals(df_2010_t['True_GDP_modeling']))
df_2010_t['Unemployment Baseline'] = df_2010_t['Unemployment Baseline'] / 100
print(df_2010_t['Unemployment Baseline'].equals(df_2010_t['Unemployment_Rate_modeling']))
```
