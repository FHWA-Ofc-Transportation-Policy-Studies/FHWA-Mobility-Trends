```python
import pandas as pd 
pd.set_option('display.max_columns', 500)

from matplotlib import pyplot as plt
```

EV

```python
EV_df = pd.read_csv('Model Data/EV Charging.csv')
EV_df = pd.melt(EV_df, id_vars = ['Row Labels', 'Scenario'], var_name = 'YEAR', value_name = 'EV Charging Count')
EV_df1, EV_df2 = EV_df[EV_df['Scenario'] == 1], EV_df[EV_df['Scenario'] == 2]

EV_df1 = EV_df1.rename(columns = {'EV Charging Count': 'EV Charging Scenario 1'})[['Row Labels', 'YEAR','EV Charging Scenario 1' ]]
EV_df2 = EV_df2.rename(columns = {'EV Charging Count': 'EV Charging Scenario 2'})[['Row Labels', 'YEAR','EV Charging Scenario 2' ]]

EV_df = EV_df1.merge(EV_df2, how = 'inner', on = ['Row Labels', 'YEAR'])
EV_df = EV_df.rename(columns={'Row Labels': 'FIPS'})
EV_df['FIPS'] = EV_df['FIPS'].astype('str')
EV_df
```

Telework

```python
tele_df = pd.read_csv('Model Data/Telework.csv')
tele_df = pd.melt(tele_df, id_vars = ['Row Labels', 'Scenario'], var_name = 'YEAR', value_name = 'Telework')

tele_df1, tele_df2, tele_df3, tele_df4 = tele_df[tele_df['Scenario'] == 'Baseline'], tele_df[tele_df['Scenario'] == 'High'], tele_df[tele_df['Scenario'] == 'Low'], tele_df[tele_df['Scenario'] == 'Mixed']
tele_df1 = tele_df1.rename(columns = {'Telework':'Telework Baseline'})[['Row Labels', 'YEAR','Telework Baseline' ]]
tele_df2 = tele_df2.rename(columns = {'Telework':'Telework High'})[['Row Labels', 'YEAR','Telework High' ]]
tele_df3 = tele_df3.rename(columns = {'Telework':'Telework Low'})[['Row Labels', 'YEAR','Telework Low' ]]
tele_df4 = tele_df4.rename(columns = {'Telework':'Telework Mixed'})[['Row Labels', 'YEAR','Telework Mixed' ]]

tele_df = tele_df1.merge(tele_df2, how = 'inner', on = ['Row Labels', 'YEAR'])
tele_df = tele_df.merge(tele_df3, how = 'inner', on = ['Row Labels', 'YEAR'])
tele_df = tele_df.merge(tele_df4, how = 'inner', on = ['Row Labels', 'YEAR'])
tele_df = tele_df.rename(columns={'Row Labels': 'FIPS'})
tele_df['FIPS'] = tele_df['FIPS'].astype('str')


tele_df
```

Unemployment

```python
un_df = pd.read_csv('Model Data/Unemployment.csv')
un_df = pd.melt(un_df.drop(columns=['FIPS']), id_vars = ['FIPS_Txt', 'Scenario'], var_name = 'YEAR', value_name = 'Unemployment')

un_df1, un_df2, un_df3  = un_df[un_df['Scenario'] == 'High'], un_df[un_df['Scenario'] == 'Low'], un_df[un_df['Scenario'] == 'Baseline']
un_df1 = un_df1.rename(columns = {'Unemployment':'Unemployment High'})[['FIPS_Txt', 'YEAR','Unemployment High']]
un_df2 = un_df2.rename(columns = {'Unemployment':'Unemployment Low'})[['FIPS_Txt', 'YEAR','Unemployment Low']]
un_df3 = un_df3.rename(columns = {'Unemployment':'Unemployment Baseline'})[['FIPS_Txt', 'YEAR','Unemployment Baseline']]

un_df = un_df1.merge(un_df2,  how = 'inner', on = ['FIPS_Txt', 'YEAR'])
un_df = un_df.merge(un_df3,  how = 'inner', on = ['FIPS_Txt', 'YEAR'])
un_df = un_df.rename(columns={'FIPS_Txt': 'FIPS'})
un_df['FIPS'] = un_df['FIPS'].astype('str')
un_df
```

Lane Miles

```python
lm_df = pd.read_csv('Model Data/Lane Miles.csv')
lm_df = pd.melt(lm_df.drop(columns=['GEMS_Typo']), id_vars = ['Row Labels', 'Scenario'], var_name = 'YEAR', value_name = 'Lane Miles')

lm_df1, lm_df2, lm_df3= lm_df[lm_df['Scenario'] == 'Baseline'], lm_df[lm_df['Scenario'] == 'Upper'], lm_df[lm_df['Scenario'] == 'Lower']
lm_df1 = lm_df1.rename(columns = {'Lane Miles':'Lane Miles Baseline'})[['Row Labels', 'YEAR','Lane Miles Baseline' ]]
lm_df2 = lm_df2.rename(columns = {'Lane Miles':'Lane Miles Upper'})[['Row Labels', 'YEAR','Lane Miles Upper' ]]
lm_df3 = lm_df3.rename(columns = {'Lane Miles':'Lane Miles Lower'})[['Row Labels', 'YEAR','Lane Miles Lower' ]]

lm_df = lm_df1.merge(lm_df2, how = 'inner', on = ['Row Labels', 'YEAR'])
lm_df = lm_df.merge(lm_df3, how = 'inner', on = ['Row Labels', 'YEAR'])
lm_df = lm_df.rename(columns={'Row Labels': 'FIPS'})
lm_df['FIPS'] = lm_df['FIPS'].astype('str')

lm_df
```

GDP

```python
gdp_df = pd.read_csv('Model Data/GDP.csv')
gdp_df = pd.melt(gdp_df.drop(columns=['FIPS_Txt', 'State_FIPS']), id_vars = ['Row Labels', 'Scenario', 'EIA_Region'], var_name = 'YEAR', value_name = 'GDP')

gdp_df1, gdp_df2, gdp_df3, gdp_df4 = gdp_df[gdp_df['Scenario'] == 'EIA Real'], gdp_df[gdp_df['Scenario'] == 'EIA Real Low'], gdp_df[gdp_df['Scenario'] == 'EIA Real High'], gdp_df[gdp_df['Scenario'] == 'Linear']
gdp_df1 = gdp_df1.rename(columns = {'GDP':'EIA REAL'})[['Row Labels', 'YEAR','EIA REAL', 'EIA_Region' ]]
gdp_df2 = gdp_df2.rename(columns = {'GDP':'EIA Real Low'})[['Row Labels', 'YEAR','EIA Real Low', 'EIA_Region' ]]
gdp_df3 = gdp_df3.rename(columns = {'GDP':'EIA Real High'})[['Row Labels', 'YEAR','EIA Real High', 'EIA_Region' ]]
gdp_df4 = gdp_df4.rename(columns = {'GDP':'Linear'})[['Row Labels', 'YEAR','Linear', 'EIA_Region' ]]

gdp_df = gdp_df1.merge(gdp_df2, how = 'inner', on = ['Row Labels', 'YEAR', 'EIA_Region'])
gdp_df = gdp_df.merge(gdp_df3, how = 'inner', on = ['Row Labels', 'YEAR', 'EIA_Region'])
gdp_df = gdp_df.merge(gdp_df4, how = 'inner', on = ['Row Labels', 'YEAR', 'EIA_Region'])
gdp_df = gdp_df.rename(columns={'Row Labels': 'FIPS'})
gdp_df['FIPS'] = gdp_df['FIPS'].astype('str')

gdp_df

```

POP

```python
pop_df = pd.read_csv('Model Data/Population.csv')
pop_df = pd.melt(pop_df.drop(columns=['FIPS']), id_vars = ['FIPS_Txt', 'Scenario'], var_name = 'YEAR', value_name = 'Pop')

pop_df1, pop_df2, pop_df3 = pop_df[pop_df['Scenario'] == 'Linear'], pop_df[pop_df['Scenario'] == 'No Loss'], pop_df[pop_df['Scenario'] == 'Cagr 10 Year']
pop_df1 = pop_df1.rename(columns = {'Pop':'Pop Linear'})[['FIPS_Txt', 'YEAR','Pop Linear', ]]
pop_df2 = pop_df2.rename(columns = {'Pop':'Pop No Loss'})[['FIPS_Txt', 'YEAR','Pop No Loss', ]]
pop_df3 = pop_df3.rename(columns = {'Pop':'Pop Cagr 10 Year'})[['FIPS_Txt', 'YEAR','Pop Cagr 10 Year', ]]

pop_df = pop_df1.merge(pop_df2, how = 'inner', on =  ['FIPS_Txt', 'YEAR'])
pop_df = pop_df.merge(pop_df3, how = 'inner', on =  ['FIPS_Txt', 'YEAR'])
pop_df = pop_df.rename(columns={'FIPS_Txt': 'FIPS'})
pop_df['FIPS'] = pop_df['FIPS'].astype('str')

pop_df
```

Join Together

```python
final_df = pop_df.merge(gdp_df, how = 'outer', on = ['FIPS', 'YEAR'])
final_df = final_df.merge(EV_df, how = 'outer', on = ['FIPS', 'YEAR'])
final_df = final_df.merge(lm_df, how = 'outer', on = ['FIPS', 'YEAR'])
final_df = final_df.merge(tele_df, how = 'outer', on = ['FIPS', 'YEAR'])
final_df = final_df.merge(un_df, how = 'outer', on = ['FIPS', 'YEAR'])
final_df = final_df[final_df['FIPS'] != 'Grand Total']
final_df = final_df[final_df['FIPS'] != '(blank)']

final_df
```

```python
percent_missing = final_df.isnull().sum() * 100 / len(final_df)
missing_value_df = pd.DataFrame({'column_name': final_df.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
final_df.to_csv('forecasting_data.csv', index = False)
```
