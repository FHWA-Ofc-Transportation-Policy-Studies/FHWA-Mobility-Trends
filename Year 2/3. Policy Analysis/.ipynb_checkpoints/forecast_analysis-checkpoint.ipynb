```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
forecast_df = pd.read_csv("C:/Users/zapate/Desktop/FHWA_Mobility_Trends/Year 2 County Level/forecasting/forecasting_data.csv")
forecast_df.head()
```

```python
population_group = forecast_df[['YEAR', 'Pop']].groupby('YEAR').sum('Pop')
plt.plot(population_group)
```

```python
gdp_group = forecast_df[['YEAR', 'EIA REAL']].groupby('YEAR').sum('EIA REAL')
plt.plot(gdp_group)
```

```python
unemployment_group = forecast_df[['YEAR', 'Unemployment Low']].groupby('YEAR').mean('Unemployment Low')
plt.plot(unemployment_group)
plt.title("AVG unemployment over time")
```
