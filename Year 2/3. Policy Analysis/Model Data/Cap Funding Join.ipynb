```python
import pandas as pd
```

```python
def melt_df(df, data_name):
    years = ['2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025',
       '2026', '2027', '2028', '2029', '2030', '2031', '2032', '2033', '2034',
       '2035', '2036', '2037', '2038', '2039', '2040', '2041', '2042', '2043',
       '2044', '2045', '2046', '2047', '2048', '2049', '2050']
    
    melted_df = pd.melt(df, id_vars = 'FIPS', value_vars = years, ignore_index= True).reset_index()
    melted_df = melted_df.rename(columns={'variable':'YEAR', 'value':data_name}).drop(columns='index')
    return(melted_df)
```

```python
df = pd.read_csv('Forecast_Data 1_2_2024.csv')
```

```python
cap_fund = pd.read_csv('Capital Funding.csv')
cap_fund = cap_fund.rename(columns={'Row Labels':'FIPS'})
cap_fund = melt_df(cap_fund, 'capital_transit_funding')
cap_fund['capital_transit_funding'] = (pd.to_numeric(cap_fund['capital_transit_funding'], errors= 'coerce').fillna(0))
cap_fund = cap_fund[cap_fund['FIPS'] != 'Grand Total']
cap_fund = cap_fund[cap_fund['FIPS'] != '(blank)']
cap_fund
```

```python
cap_fund['FIPS'] = cap_fund['FIPS'].astype(int)
cap_fund['YEAR'] = cap_fund['YEAR'].astype(int)
```

```python
df = pd.merge(df, cap_fund, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
modeling_dataset = pd.read_csv('C:/Users/zapate/Desktop/FHWA_Mobility_Trends/Year 2 County Level/Modeling_Year1_County_level/Data/County_Year2_1_10_24.csv')
modeling_dataset = modeling_dataset[modeling_dataset['YEAR'] == 2020]
modeling_dataset = modeling_dataset[['Full_FIPS_Code', 'ALAND_SQMI']].drop_duplicates()
modeling_dataset = modeling_dataset.rename(columns = {'Full_FIPS_Code':'FIPS'})
modeling_dataset = modeling_dataset.dropna()
modeling_dataset.sort_values('FIPS')
```

```python
df = pd.merge(df, modeling_dataset, how = 'left', on = [ 'FIPS'])
df['ALAND_SQMI'] = (pd.to_numeric(df['ALAND_SQMI'], errors='coerce').fillna(0))
```

```python
df['POPDENSITY'] = df['Pop Linear'] / df['ALAND_SQMI']
df['POPDENSITY'] = (pd.to_numeric(df['POPDENSITY'], errors='coerce').fillna(0))
```

```python
df.drop_duplicates().to_csv('Forecast_Data 1_16_2024.csv', index = False)
```
