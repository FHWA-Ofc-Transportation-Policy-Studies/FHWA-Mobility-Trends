Creating population desnity forecast for 2022 - 2050

County Population density = County Population / County area size (sq miles)

Using population forecasts provided by Matt Jacobs 

Assuming county area size stays the same as it is in 2020. 

```python
import os
import pandas as pd 
```

```python
#loading in county area data 
datapath = 'C:/Users/hrowe/Documents/FHWA mobility trend report/T4 - Forecasting/Year 2/FHWA_Mobility_Trends/Year 2 County Level/Data_engineering_Year1_County_level/Year2Indicators/etl/year2_indicators'
area_df = pd.read_csv(datapath + '/county_area.csv', index_col = 0)
area_df
```

```python
#loading in population forecast data 
datapath = os.getcwd()
pop_no_loss_df = pd.read_csv(datapath + '/Population No Loss.csv')
pop_no_loss_df 
```

```python
def melt_df(df, data_name):
    years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025',
       '2026', '2027', '2028', '2029', '2030', '2031', '2032', '2033', '2034',
       '2035', '2036', '2037', '2038', '2039', '2040', '2041', '2042', '2043',
       '2044', '2045', '2046', '2047', '2048', '2049', '2050']
    
    melted_df = pd.melt(df, id_vars = 'FIPS', value_vars = years, ignore_index= True).reset_index()
    melted_df = melted_df.rename(columns={'variable':'YEAR', 'value':data_name}).drop(columns='index')
    return(melted_df)
```

```python
pop_no_loss_df = melt_df(pop_no_loss_df, 'pop_no_loss')
pop_no_loss_df
```

```python
pop_no_loss_df.groupby('YEAR').pop_no_loss.sum().plot().set(title = "Sum of county Population No Loss Forecast")
```

```python
pop_no_loss_df['YEAR'] = pd.to_numeric(pop_no_loss_df['YEAR'])
pop_no_loss_df['pop_no_loss'] = pd.to_numeric(pop_no_loss_df['pop_no_loss'], errors='coerce')
area_df = area_df.rename(columns={'GEOID': 'FIPS'})
```

```python
pop_no_loss_df.dtypes
```

```python
area_df.dtypes
```

```python
df_pop_density = pd.merge(pop_no_loss_df,area_df, on = ['FIPS', 'YEAR'] , how='left')
df_pop_density
```

```python
#set land area equal to its previous value for all cases of missing land area values (assuming land areas stays the same year to year for each county)
df_pop_density_for = df_pop_density
df_pop_density_for['ALAND_SQMI'] = df_pop_density_for.groupby(['FIPS'])['ALAND_SQMI'].fillna(method='ffill')
```

```python
#create population density measure from 2000 - 2050
df_pop_density_for['POP_DENSITY_NO_LOSS'] = df_pop_density_for['pop_no_loss'] / df_pop_density_for['ALAND_SQMI']
```

```python
#checking sum of national county land area, expecting stable numbers from 2000 - 2050 
df_pop_density_for.groupby('YEAR').ALAND_SQMI.sum().plot().set(title = "Sum of county ALAND_SQMI")
```

```python
#only a 0.4% increase in county area between 2000 census and 2010 census numbers, seems reasoable and not a cause for concern
(3.5175e6 - 3.5325e6 ) /3.5175e6
```

```python
df_pop_density_for.groupby('YEAR').POP_DENSITY_NO_LOSS.median().plot().set(title = "Median of national population density")
```
