```python
import pandas as pd
```

```python
def melt_df(df, data_name):
    years = ['2005', '2006', '2007',
       '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025',
       '2026', '2027', '2028', '2029', '2030', '2031', '2032', '2033', '2034',
       '2035', '2036', '2037', '2038', '2039', '2040', '2041', '2042', '2043',
       '2044', '2045', '2046', '2047', '2048', '2049', '2050']
    
    melted_df = pd.melt(df, id_vars = 'FIPS', value_vars = years, ignore_index= True).reset_index()
    melted_df = melted_df.rename(columns={'variable':'YEAR', 'value':data_name}).drop(columns='index')
    return(melted_df)
```

```python
#population
df_pop_lin = melt_df(pd.read_csv('Population Linear.csv'), 'Pop Linear')
df_pop_cagr = melt_df(pd.read_csv('Pop 10 Year CAGR.csv'), 'Pop 10 CAGR')
df_pop_no_loss = melt_df(pd.read_csv('Population No Loss.csv'), 'Pop No Loss')

df = pd.merge(df_pop_lin, df_pop_cagr, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_pop_no_loss, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#transitn upt
df_tran = melt_df(pd.read_csv('Transit CAGR.csv'), 'Transit')

df = pd.merge(df, df_tran, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#EV Charging
df_ev1 = melt_df(pd.read_csv('Charging Station 1.csv'), 'EV 1')
df_ev2 = melt_df(pd.read_csv('Charging Station 2.csv'), 'EV 2')

df = pd.merge(df, df_ev1, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_ev2, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#GDP
df_gdp = melt_df(pd.read_csv('EIA Real GDP.csv'), 'Real GDP')
df_gdplow = melt_df(pd.read_csv('EIA Real GDP High.csv'), 'Real GDP Low')
df_gdphigh = melt_df(pd.read_csv('EIA Real GDP Low.csv'), 'Real GDP High')

df = pd.merge(df, df_gdp, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_gdplow, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_gdphigh, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#lane miles
df_lm1 = melt_df(pd.read_csv('Lane Miles 1.csv'), 'Lane Miles 1')
df_lm2 = melt_df(pd.read_csv('Lane Miles 2.csv'), 'Lane Miles 2')
df_lm3 = melt_df(pd.read_csv('Lane Miles 3.csv'), 'Lane Miles 3')

df = pd.merge(df, df_lm1, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_lm2, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_lm3, how = 'outer', on = ['YEAR', 'FIPS'])
df
```

```python
#Telework
df_tele1 = melt_df(pd.read_csv('Telework 1.csv'), 'Telework 1')
df_tele2 = melt_df(pd.read_csv('Telework 2.csv'), 'Telework 2')
df_tele3 = melt_df(pd.read_csv('Telework 3.csv'), 'Telework 3')
df_tele4 = melt_df(pd.read_csv('Telework 4.csv'), 'Telework 4')

df = pd.merge(df, df_tele1, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_tele2, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_tele3, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_tele4, how = 'outer', on = ['YEAR', 'FIPS'])
df
```

```python
#Unemployment
df_unemp_ma = melt_df(pd.read_csv('Unemployment MA.csv'), 'Unemployment MA')
df_unemp_high = melt_df(pd.read_csv('Unemployment High Econ.csv'), 'Unemployment High Econ')
df_unemp_low = melt_df(pd.read_csv('Unemployment Low Econ.csv'), 'Unemployment Low Econ')

df = pd.merge(df, df_unemp_ma, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_unemp_high, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_unemp_low, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#courier revenue
df_courier_base = melt_df(pd.read_csv('Courier Forecast Baseline.csv'), 'Courier Baseline')
df_courier_High = melt_df(pd.read_csv('Courier Forecast High.csv'), 'Courier High')
df_courier_Low = melt_df(pd.read_csv('Courier Forecast Low.csv'), 'Courier Low')

df = pd.merge(df, df_courier_base, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_courier_High, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_courier_Low, how = 'outer', on = ['YEAR', 'FIPS'])
```

```python
#Driver rev
df_driver_base = melt_df(pd.read_csv('Driver Rev Forecast .csv'), 'Driver Rev')
df_driver_Baseline = melt_df(pd.read_csv('Driver Rev Baseline.csv'), 'Driver Rev Baseline')
df_driver_High = melt_df(pd.read_csv('Driver Rev High.csv'), 'Driver Rev High')
df_driver_Low = melt_df(pd.read_csv('Driver Rev Low.csv'), 'Driver Rev Low')

df = pd.merge(df, df_driver_base, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_driver_Baseline, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_driver_High, how = 'outer', on = ['YEAR', 'FIPS'])
df = pd.merge(df, df_driver_Low, how = 'outer', on = ['YEAR', 'FIPS'])
```
