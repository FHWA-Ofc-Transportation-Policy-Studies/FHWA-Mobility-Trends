### 1. Are variables correlated?

```python
#function to plot spearman  correlation matrix 
def corr_matrix(data, corr_col, chart_title):
    data = data.loc[:, data.columns.intersection(corr_col)]
    plt.figure(figsize=(15,10))
    sns.heatmap(data.corr(method = 'spearman').round(2),cmap='Blues',annot=True) 
    plt.title(chart_title, y=1.02)
    
#example call of corr_matrix function
plt.show(corr_matrix(data = df, corr_col = ['VMT','TOTAL_EMISSIONS', 'POP_DENSITY', 'LNMILES', 'COURIER_EMPLY'], chart_title = "Correlation Matrix: All US Counties" ))
```

### 2. Are variables normally distributed? or skewed?

```python
#function that plots distriubtion (density) for each variable in a dataset 
import warnings
def dist_plots(data, col_num, chart_title, x_axis):
    warnings.filterwarnings("ignore")
    l = data.columns.values
    number_of_columns = col_num
    number_of_rows = int(len(l)-1/number_of_columns)
    plt.figure(figsize=(5*number_of_columns,5*number_of_rows))
    for i in range(0,len(l)):
        plt.subplot(number_of_rows + 1,number_of_columns,i+1)
        sns.distplot(data[l[i]],kde=True)
        
        if (x_axis == 'zoom'):
            plt.xlim(0, 0.2 * (data[l[i]].max()) )
        if (x_axis == 'log'):
            plt.xscale('log')

    plt.suptitle(chart_title, x = 0.5, y = 0.90)
```

```python
#example calling dist_plots function
df_ondemand = df[['VMT','POPULATION', 'True_GDP', 'Unemployment_Rate', 'LNMILES', 'DRIVER_NONEMP_RCPTOT_REAL', 'INTERNET_CONNECTIONS', 'HH_ZERO_VEH']]
plt.show(dist_plots(df_ondemand, 3, "Distribution Plots", x_axis = 'zoom'))
```

### 3. How many null values exist for a variable? Where (what regions and years) do the null values come from?

```python
nulls = pd.DataFrame(df.isna().sum(), columns = ['Count of NAs']).reset_index()
nulls
```

```python
#function shows the missing and non missing values for a variable by a specified grouping 
def variable_nulls(data, var, group):
    df_nulls = pd.DataFrame(data[data[str(var)].isna()].groupby(str(group)).size(), columns = ['Missing Obs']).reset_index()
    df_non_nulls = pd.DataFrame(data[data[str(var)].notna()].groupby(str(group)).size(), columns = ['Non Missing Obs']).reset_index()
    df_merge = pd.merge(df_nulls, df_non_nulls, on = [str(group)] , how='outer')
    df_merge['Total Obs'] =  df_merge['Missing Obs'] + df_merge['Non Missing Obs']
    df_merge['Percent Missing'] = round(df_merge['Missing Obs'] / ( df_merge['Missing Obs'] + df_merge['Non Missing Obs'] ), 2)
    df_merge = df_merge.sort_values(by = str(group), ascending = True)
    return(df_merge)

#example 'variable_nulls' function call
variable_nulls(df, 'WAREHOUSE_EMPLY', 'STATE_CODE')
```

### 4. What does the aggergate national time series look like for each variable?

```python
def plot_national_agg (df, var, title):
    #df.groupby('YEAR')[str(var)].sum().plot().set(title = str(title))
    plt.plot(df.groupby('YEAR')[str(var)].sum())
    plt.ylabel(str(var))
    plt.xticks(range(2000,2021,2))
    plt.title(str(title))
    

plot_national_agg(df, var = 'INTERNET_CONNECTIONS', title = "Sum of county INTERNET_CONNECTIONS")
```
