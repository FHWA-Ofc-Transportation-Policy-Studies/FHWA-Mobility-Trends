##### Function to perform sq root transformation on any column

```python
def sqrt_trans(data, col_name):
    new_col_name = col_name + '_sqrt'
    data[new_col_name] = np.sqrt(data[col_name])
```

```python
sqrt_trans(df_all_pm, 'VMT')
sqrt_trans(df_all_pm, 'POPULATION')
sqrt_trans(df_all_pm, 'TOTAL_EMISSIONS')
sqrt_trans(df_all_pm, 'True_GDP')
sqrt_trans(df_all_pm, 'Charging_Stations')
sqrt_trans(df_all_pm, 'LNMILES')
sqrt_trans(df_all_pm, 'distributed_UPT')
```

##### Function to create lag1 column for any existing column 

```python
def lag_col(data, col_name, n_lag):
    #create name for the lag column
    new_col_name = col_name + '_lag' + str(n_lag)
    
    #sort the data by year, group by fips code, then take the lag of the specificed 'col_name' 
    data[new_col_name] = data.sort_values(by = ['YEAR']).groupby('True_Fips')[col_name].shift(n_lag)
    return data
```

```python
df = lag_col(df_all_pm, 'VMT', 1)
df = lag_col(df_all_pm, 'TOTAL_EMISSIONS', 1)
df = lag_col(df_all_pm, 'TOTAL_EMISSIONS', 2)
df
```
