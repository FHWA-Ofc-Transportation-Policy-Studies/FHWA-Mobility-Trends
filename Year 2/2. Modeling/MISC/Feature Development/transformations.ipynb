```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
import numpy as np
```

```python
df = pd.read_csv('df_7_31_w_sqrt_transformations.csv')
df
```

```python
check_skew1 = ['VMT', 'POPULATION',
       'TOTAL_EMISSIONS', 'True_GDP',  'Unemployment_Rate',
       'Charging_Stations', 'LNMILES', 'TMS', 'TELEWORK', 'distributed_UPT',
        ]

check_skew2 = ['POPULATION_sqrt', 'TOTAL_EMISSIONS_sqrt',
       'VMT_sqrt', 'True_GDP_sqrt', 'Charging_Stations_sqrt', 'LNMILES_sqrt',
       'distributed_UPT_sqrt']


print('Current Skew')

for i in check_skew1:
    print(i, ':', df[i].skew())

print('cube root transformation')
for i in check_skew1:
    print(i, ':', np.cbrt(df[i].skew()))


```

```python
print('log transformation')
for i in check_skew1:
    print(i, ':', np.log(df[i].skew()))
```

```python
sn.distplot(df['VMT'])
plt.title('VMT w/o transformation')
```

```python
sn.kdeplot(np.log(df['VMT']))
plt.title('VMT w/ log transformation')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
df_vmt = df[['VMT']]
df_vmt = df_vmt.loc[~(df_vmt==0).all(axis=1)]
fig.suptitle('No transformation vs. Log Transformation - VMT')
sn.distplot(df_vmt['VMT'], ax = ax1)
sn.distplot(np.log(df_vmt['VMT'] + 1), ax = ax2)
```

```python
def dist_plot_indicators(indicator):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    df_vmt = df[[indicator]]
    df_vmt = df_vmt.loc[~(df_vmt==0).all(axis=1)]
    fig.suptitle('No transformation vs. Log Transformation - ' + indicator)
    sn.distplot(df_vmt[indicator], ax = ax1)
    sn.distplot(np.log(df_vmt[indicator] + 1), ax = ax2)
```
